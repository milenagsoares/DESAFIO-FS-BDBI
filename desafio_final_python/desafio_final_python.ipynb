```python
#Plote um gráfico de barras (bar) que exiba a quantidade de
#funcionários em cada setor (coluna Department);
# Plot
department = df.groupby('Department').size()
department.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Quantidade de Funcionários por Setor')
plt.xlabel('Setor')
plt.ylabel('Número de Funcionários')
plt.show()
```

```python
#Plote um gráfico de linha (line) que exiba a quantidade de
#funcionários para cada nível de escolaridade (coluna Education); 

education = df.groupby('Education').size()

education.plot(kind='line', marker='*', color='blue')

plt.title('Quantidade de Funcionários por Nível de Escolaridade')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Número de Funcionários')

plt.show()
```

```python
attrition = df['Attrition'].value_counts()
plt.figure(figsize=(8, 6))
attrition.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['green', 'red'], wedgeprops={'edgecolor': 'black'})
plt.title('Distribuição de Funcionários por Atrito na Empresa')
plt.show()
```

```python
sns.histplot(data=df, x='Age', kde=True)
plt.title('Distribuição de Idade dos Funcionários')
plt.show()
```

```python
sns.boxplot(data=df, x='JobRole', y='MonthlyIncome')
plt.title('Distribuição de Renda Mensal por Cargo')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.catplot(data=df, x='EducationField', y='MonthlyIncome', kind='bar')
plt.title('Média de Renda Mensal por Campo de Educação')
plt.xticks(rotation=45)
plt.show()

```
